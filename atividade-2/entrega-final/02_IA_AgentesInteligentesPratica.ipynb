```python
# Classes herdadas a partir da classe Thing
class Comida(Thing):
    pass

class Lago(Thing):
    pass

class Arvore(Thing):
    pass

class Pessoa(Thing):
    pass

class Agua(Thing):
    pass

class Rato(Thing):
    pass

class Grama(Thing):
    pass

class Obstaculo(Thing):
    pass

class Cachorro(Thing):
    pass

class Gato(Thing):
    pass

class ParqueDoCarmo(Environment):
    def percept(self, agent):
        '''Imprime e retorna uma lista de coisas que estão na localização do nosso agente'''
        things = self.list_things_at(agent.location)
        print(things)
        return things
    
    def execute_action(self, agent, action):
        '''Altera o estado do ambiente com base no que o agente faz.'''
        if action == "move":
            agent.move()
        elif action == "come":
            items = self.list_things_at(agent.location, tclass=Comida)
            if len(items) != 0:
                if agent.come(items[0]): # Fazer a escolha do gato comer o primeiro item
                    self.delete_thing(items[0]) # Delete-o do parque depois.
        elif action == "nada":
            items = self.list_things_at(agent.location, tclass=Lago)
            if len(items) != 0:
                if agent.nada(items[0]): # Fazer a escolha do gato nadar no primeiro item
                    self.delete_thing(items[0]) # Delete-o do parque depois.
        elif action == "sobe":
            items = self.list_things_at(agent.location, tclass=Arvore)
            if len(items) != 0:
                if agent.sobe(items[0]): # Fazer a escolha do gato subir no primeiro item
                    self.delete_thing(items[0]) # Delete-o do parque depois.
        elif action == "acaricia":
            items = self.list_things_at(agent.location, tclass=Pessoa)
            if len(items) != 0:
                if agent.acaricia(items[0]): # Fazer a escolha do gato acariciar o primeiro item
                    self.delete_thing(items[0]) # Delete-o do parque depois.
        elif action == "bebe":
            items = self.list_things_at(agent.location, tclass=Agua)
            if len(items) != 0:
                if agent.bebe(items[0]): # Fazer a escolha do gato beber o primeiro item
                    self.delete_thing(items[0]) # Delete-o do parque depois.
        elif action == "caca":
            items = self.list_things_at(agent.location, tclass=Rato)
            if len(items) != 0:
                if agent.caca(items[0]): # Fazer a escolha do gato caçar o primeiro item
                    self.delete_thing(items[0]) # Delete-o do parque depois.
        elif action == "rola":
            items = self.list_things_at(agent.location, tclass=Grama)
            if len(items) != 0:
                if agent.rola(items[0]): # Fazer a escolha do gato rolar no primeiro item
                    self.delete_thing(items[0]) # Delete-o do parque depois.
        elif action == "pula":
            items = self.list_things_at(agent.location, tclass=Obstaculo)
            if len(items) != 0:
                if agent.pula(items[0]): # Fazer a escolha do gato pular o primeiro item
                    self.delete_thing(items[0]) # Delete-o do parque depois.
        elif action == "foge":
            items = self.list_things_at(agent.location, tclass=Cachorro)
            if len(items) != 0:
                if agent.foge(items[0]): # Fazer a escolha do gato fugir o primeiro item
                    self.delete_thing(items[0]) # Delete-o do parque depois.
        elif action == "namora":
            items = self.list_things_at(agent.location, tclass=Gato)
            if len(items) != 0:
                if agent.namora(items[0]): # Fazer a escolha do gato namorar o primeiro item
                    self.delete_thing(items[0]) # Delete-o do parque depois.
                    
    def is_done(self):
        '''Por padrão, o ambiente termina quando não podemos encontrar um agente vivo,
         mas para evitar matar o nosso gato robô, o ambiente termina apenas quando não há mais comida ou água'''
        no_edibles = not any(isinstance(thing, Comida) or isinstance(thing, Lago) or isinstance(thing, Arvore) or isinstance(thing, Pessoa)  or isinstance(thing, Agua)  or isinstance(thing, Rato)  or isinstance(thing, Grama)  or isinstance(thing, Obstaculo)  or isinstance(thing, Cachorro)  or isinstance(thing, Gato) for thing in self.things)
        dead_agents = not any(agent.is_alive() for agent in self.agents)
        return dead_agents or no_edibles
```

# Programa #
Agora que temos uma classe ParqueDoCarmo, precisamos implementar um módulo programa para o nosso gato. Um programa controla como o gato age sobre seu ambiente. Nosso programa será muito simples e é mostrado na tabela abaixo.

```python
# Cria uma classe a partir do módulo agente
class RoboGato(Agent):
    location = 1
    
    def move(self):
        self.location += 1
        
    def come(self, thing):
        '''Retorna True em caso de sucesso ou False caso contrário'''
        if isinstance(thing, Comida):
            print("Robô Gato: Comeu às {}.".format(self.location))
            return True
        return False

    def nada(self, thing):
        '''Retorna True em caso de sucesso ou False caso contrário'''
        if isinstance(thing, Lago):
            print("Robô Gato: Nadou às {}.".format(self.location))
            return True
        return False

    def sobe(self, thing):
        '''Retorna True em caso de sucesso ou False caso contrário'''
        if isinstance(thing, Arvore):
            print("Robô Gato: Subiu às {}.".format(self.location))
            return True
        return False

    def acaricia(self, thing):
        '''Retorna True em caso de sucesso ou False caso contrário'''
        if isinstance(thing, Pessoa):
            print("Robô Gato: Fez carinho às {}.".format(self.location))
            return True
        return False

    def bebe(self, thing):
        '''Retorna True em caso de sucesso ou False caso contrário'''
        if isinstance(thing, Agua):
            print("Robô Gato: Bebeu às {}.".format(self.location))
            return True
        return False
    
    def caca(self, thing):
        '''Retorna True em caso de sucesso ou False caso contrário'''
        if isinstance(thing, Rato):
            print("Robô Gato: Caçou às {}.".format(self.location))
            return True
        return False

    def rola(self, thing):
        '''Retorna True em caso de sucesso ou False caso contrário'''
        if isinstance(thing, Grama):
            print("Robô Gato: Rolou na grama às {}.".format(self.location))
            return True
        return False

    def pula(self, thing):
        '''Retorna True em caso de sucesso ou False caso contrário'''
        if isinstance(thing, Obstaculo):
            print("Robô Gato: Pulou às {}.".format(self.location))
            return True
        return False

    def foge(self, thing):
        '''Retorna True em caso de sucesso ou False caso contrário'''
        if isinstance(thing, Cachorro):
            print("Robô Gato: Fugiu às {}.".format(self.location))
            return True
        return False

    def namora(self, thing):
        '''Retorna True em caso de sucesso ou False caso contrário'''
        if isinstance(thing, Gato):
            print("Robô Gato: Namorou às {}.".format(self.location))
            return True
        return False
            
def program(percepts):
    '''Retorna uma ação baseada em suas percepções'''
    for p in percepts:
        if isinstance(p, Comida):
            return 'come'
        elif isinstance(p, Lago):
            return 'nada'
        elif isinstance(p, Arvore):
            return 'sobe'
        elif isinstance(p, Pessoa):
            return 'acaricia'
        elif isinstance(p, Agua):
            return 'bebe'
        elif isinstance(p, Rato):
            return 'caca'
        elif isinstance(p, Grama):
            return 'rola'
        elif isinstance(p, Obstaculo):
            return 'pula'
        elif isinstance(p, Cachorro):
            return 'foge'
        elif isinstance(p, Gato):
            return 'namora'
    return 'move'
```

```python
parque = ParqueDoCarmo()
gato = RoboGato(program)
```
